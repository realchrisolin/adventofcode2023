```python
possible_game_ids = []
game_data = {}
red_cube_limit = 12
green_cube_limit = 13
blue_cube_limit = 14
```

```python
import re

def parse_game_info(data):

    # replace colon with semicolon so we can split on ; and game ID is always at index 0
    sanitized_data = data.replace(':',';')
    split_data = sanitized_data.split(';')

    # extract game ID
    game_id = split_data[0]
    match = re.match(r"Game (\d+)", game_id)
    game_id = str(match.group(1))
    split_data.pop(0)

    # split the individual rounds and store them in a dict linked to the game ID
    game_data = []
    for game in split_data:
        game_data.append(game)
    parsed_data = {
        game_id:  [game_data]
    }
    return parsed_data
```

```python
parsed_games = {}
with open('2.input', 'r') as input_file:
    for line in input_file:
        parsed_info = parse_game_info(line)
        parsed_games.update(parsed_info)
```

```python
def parse_cube_info(data):
    # store color values in a dict for easy access later
    color_matches = {}
    for cube in data.split(','):
        matches = re.findall(r"(\d+) (\w+)", cube)
        result_dict = {key: value for value, key in matches}
        color_matches.update(result_dict)
    return color_matches
```

```python
possible = True
for game_id, game_data in parsed_games.items():
    for round, round_data in enumerate(game_data[0]):
        red_cube = 0
        green_cube = 0
        blue_cube = 0
        cube_info = parse_cube_info(round_data.strip())
        if 'red' in cube_info.keys():
            red_cube = int(cube_info['red'])
        if 'green' in cube_info.keys():
            green_cube = int(cube_info['green'])
        if 'blue' in cube_info.keys():
            blue_cube = int(cube_info['blue'])
        print(f'Game: {game_id}')
        print(f'Round: {round}')
        print(f'*'* 8)
        print(f'Red cube: {red_cube}')
        print(f'Green cube: {green_cube}')
        print(f'Blue cube: {blue_cube}')
        print('#'*8)
        if (red_cube <= red_cube_limit) and (green_cube <= green_cube_limit) and (blue_cube <= blue_cube_limit):
            possible = True
            continue
        else:
            possible = False
            break
    if possible:
        possible_game_ids.append(int(game_id))
        print(f'Game {game_id} is possible.')
```

The Elf says they've stopped producing snow because they aren't getting any water! He isn't sure why the water stopped; however, he can show you how to get to the water source to check it out for yourself. It's just up ahead!

As you continue your walk, the Elf poses a second question: in each game you played, what is the fewest number of cubes of each color that could have been in the bag to make the game possible?

Again consider the example games from earlier:

Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue
Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red
Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red
Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
In game 1, the game could have been played with as few as 4 red, 2 green, and 6 blue cubes. If any color had even one fewer cube, the game would have been impossible.
Game 2 could have been played with a minimum of 1 red, 3 green, and 4 blue cubes.
Game 3 must have been played with at least 20 red, 13 green, and 6 blue cubes.
Game 4 required at least 14 red, 3 green, and 15 blue cubes.
Game 5 needed no fewer than 6 red, 3 green, and 2 blue cubes in the bag.
The power of a set of cubes is equal to the numbers of red, green, and blue cubes multiplied together. The power of the minimum set of cubes in game 1 is 48. In games 2-5 it was 12, 1560, 630, and 36, respectively. Adding up these five powers produces the sum 2286.

For each game, find the minimum set of cubes that must have been present. What is the sum of the power of these sets?

```python
import re


def parse_game_info(data):
    # replace colon with semicolon so we can split on ; and game ID is always at index 0
    sanitized_data = data.replace(':', ';')
    split_data = sanitized_data.split(';')

    # extract game ID
    game_id = split_data[0]
    match = re.match(r"Game (\d+)", game_id)
    game_id = str(match.group(1))
    split_data.pop(0)

    # split the individual rounds and store them in a dict linked to the game ID
    game_data = []
    for game in split_data:
        game_data.append(game)
    parsed_data = {
        game_id: [game_data]
    }
    return parsed_data
```

```python
parsed_games = {}
with open('2.input', 'r') as input_file:
    for line in input_file:
        parsed_info = parse_game_info(line)
        parsed_games.update(parsed_info)
```

```python
def parse_cube_info(data):
    # store color values in a dict for easy access later
    color_matches = {}
    for cube in data.split(','):
        matches = re.findall(r"(\d+) (\w+)", cube)
        result_dict = {key: value for value, key in matches}
        color_matches.update(result_dict)
    return color_matches
```

```python
possible = True
game_powers = []
for game_id, game_data in parsed_games.items():
    max_red = 0
    max_green = 0
    max_blue = 0
    for round, round_data in enumerate(game_data[0]):
        print(f'Game: {game_id}')
        print(f'Round: {round}')
        print(round_data)
        red_cube = 0
        green_cube = 0
        blue_cube = 0
        cube_info = parse_cube_info(round_data.strip())
        if 'red' in cube_info.keys():
            red_cube = int(cube_info['red'])
        if 'green' in cube_info.keys():
            green_cube = int(cube_info['green'])
        if 'blue' in cube_info.keys():
            blue_cube = int(cube_info['blue'])
        if red_cube > max_red:
            max_red = red_cube
        if green_cube > max_green:
            max_green = green_cube
        if blue_cube > max_blue:
            max_blue = blue_cube

    print(f'Game: {game_id}')
    print(f'max red: {max_red}')
    print(f'max green: {max_green}')
    print(f'max blue: {max_blue}')
    power = (max_red * max_green * max_blue)
    print(f'Power: {power}')
    game_powers.append(power)
```
